```python
df2 = pd.DataFrame(data)

# Set the order of the columns
df2 = df2[ ['Year','Birth_Ratio_%_using_Steroids','Chorioamnionitis']]
print(df2)
df2.plot(x='Year',y=['Birth_Ratio_%_using_Steroids','Chorioamnionitis'])
```

```python
#Labour charecteristics 
#With years the number of mothers requiring steroids, induction of labour, augmentation of labour, chorioamnionitis, anaesthesia is showing a general upward trend
#This might imply either people are starting to have more trust in medical technologies and hence preferring medical assistance like steroids, induction of labour
#This might also imply that people are being more prone to diseases and infections because they are leading an unhealthy lifestyle and hence requiring steroids, anaesthesia, etc
from matplotlib import pyplot as plt
df2.plot(x='Year',y=['Birth_Ratio_%_using_Steroids','Induction of labour','Augmentation of Labour','Chorioamnionitis','Anaesthesia'])
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
#Pregnancy risk factors showing a general upward trend 
#With years the ratio of number of deliveries(Yes/No) having the following diseases or requiring assistance is increasing 
#This might imply either people are having more trust in medical technologies and hence preferring medical assistance like assistive reproductive reproductive technologies and infertility treatment
#This might also imply that people are leading more prone to diseases and infections and hence requiring assistance
data2={'Year':[2016,2017,2018,2019,2020,2021],'Pre-pregnancy diabetes':[0.86,0.95,0.96,0.9762,1.0766,1.125],'Gestation diabetes':[6.34,6.8338,7.1657,7.3977,8.4976,9.029],'Pre-pregnancy Hypertension':[1.75,1.915,2.134,2.2285,2.257,2.795],'Infertility Treatment Used':[1.744,1.915,2.01,2.096,2.047,2.275],'Assistive Reproductive Technology':[15.4545,17.27272727,18.18181818,21,21,24]}
df3 = pd.DataFrame(data2)
df3
df3 = df3[ ['Year','Pre-pregnancy diabetes','Gestation diabetes','Pre-pregnancy Hypertension','Infertility Treatment Used','Assistive Reproductive Technology']]

df3.plot(x='Year',y=['Pre-pregnancy diabetes','Gestation diabetes','Pre-pregnancy Hypertension','Infertility Treatment Used','Assistive Reproductive Technology'])
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
